## what are the locations with the fastest roller coasters(minimum of 10 coaaters at that location)

```python
df.value_counts(['Location'])
```

```python
mod_df=df.query('     Location != "Other"  ')
mod_df
```

```python
mod_df['Location'].value_counts()
```

```python
mod_df=df.query('     Location != "Other"  ') \
.groupby('Location')['Speed_mph'] \
.agg(['mean','count']) \
.query("count>=10") \
.sort_values('mean') \
.reset_index()
mod_df
```

```python
mod_df['Location']
```

```python
mod_df.plot(kind='barh',
            x='Location',
            y='mean',
            xlabel='Speed_mph',
            ylabel='Location Name',
            title='the locations with the fastest roller coasters(minimum of 10 coaaters at that location)')
```

```python
mod_df.plot(kind='bar',
            x='Location',
            y='mean',
            xlabel='Location Name ',
            ylabel=' Speed_mph',
            title='the locations with the fastest roller coasters(minimum of 10 coaaters at that location)',
            figsize=(10,10)
            ,)
plt.show()
```
